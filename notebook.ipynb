```python
from random import random,shuffle
class leafGame:
    def __init__(self,players):
        self.leaves = int(10**(5*random()))
        self.hist = []
        shuffle(players)
        self.players = [{"name":p.__name__,"leaves":0,"func":p}for p in players]

    def round(self):
        if abs(self.players[0]['leaves']-self.players[1]['leaves'])>self.leaves:
            return False
        player =  self.players[len(self.hist)%len(self.players)]
        ask = player["func"](self.hist)
        if ask > self.leaves:
            self.hist.append({"asked":ask,"rewarded":False,"player":player})
        elif ask <= self.leaves:
            self.hist.append({"asked":ask,"rewarded":True,"player":player})
            self.leaves -= ask
            player['leaves'] += ask
            
        return True

    def full_game(self):
        while self.round():
            pass
        p1=self.players[0]
        p2=self.players[1]
        if p1['leaves'] > p2['leaves']:
            print(p1['name'],p1['leaves'])
            print(p2["name"],p2['leaves'])
        elif p2['leaves']>p1['leaves']:
            print(p2["name"],p2['leaves'])
            print(p1['name'],p1['leaves'])
        else:
            print("DRAW")
        return self.hist
game = leafGame([ask_one,ask_twice_the_max_approved])

game.full_game()

```
